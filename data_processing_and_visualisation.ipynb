```python
import pandas as pd

df = pd.read_csv("amazon.csv")
df = df.drop_duplicates()
df["discounted_price"] = df["discounted_price"].str.replace("₹", "").str.replace(",", "").astype(float)
df["actual_price"] = df["actual_price"].str.replace("₹", "").str.replace(",", "").astype(float)
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "").astype(float)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["rating_count"] = df["rating_count"].str.replace(",", "")
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].median())
df.head()
```

```python
df["revenue"] = df["discounted_price"] * df["rating_count"]
total_revenue = df["revenue"].sum()
print("Total Estimated Revenue:", total_revenue)
```

```python
top_products = df.groupby("product_name")["revenue"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
category_sales = df.groupby("category")["revenue"].sum().sort_values(ascending=False).head(10)
print(category_sales)
```

```python
category_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False).head(10)
print(category_rating)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(df["rating"].dropna(), bins=10)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_categories.index, top_categories.values)
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["discounted_price"].dropna(), bins=10)
plt.title("Discounted Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["rating"], df["discounted_price"])
plt.title("Rating vs Price")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.show()
```

```python
avg_price = df.groupby("category")["discounted_price"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(avg_price.index, avg_price.values)
plt.title("Average Price per Category")
plt.xlabel("Category")
plt.ylabel("Avg Price")
plt.xticks(rotation=90)
plt.show()
```

## Summary of the dataset

The dataset contains 1465 products and 16 features including price, rating, and reviews.
It is mainly an Amazon product catalog dataset used for sales and review analysis.
Most products have ratings clustered between 4.0 and 4.5, indicating generally good reviews.
The average rating is ~4.10, showing positive customer satisfaction overall.
The average discounted price is around ₹3125, but prices vary widely.
A few products have extremely high prices, creating a right-skewed price distribution.
The dataset is suitable for business insights like category performance and pricing strategy.
