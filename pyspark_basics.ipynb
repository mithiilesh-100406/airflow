```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Simple ETL Pipeline") \
    .getOrCreate()
```

```python
data = [(1, "Alice", 5000),(2, "Bob", 7000),(3, "Mithilesh", 6000),(4, "Badri", 8000),(5, "Arivu", 9000)]
columns = ["id", "name", "salary"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col

df_transformed = df \
    .withColumn("bonus", col("salary") * 0.1) \
    .filter(col("salary") > 6000)

df_transformed.show()
```

```python
df_transformed.write.mode("overwrite").parquet("output/salary_data")
```

```python
df = spark.read.parquet("output/salary_data")
df.show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("DataFrame Operations") \
    .getOrCreate()

employees = [(1, "Alice", "IT", 7000),(2, "Bob", "HR", 5000),(3, "Mithilesh", "IT", 6000),(4, "Badri", "Finance", 8000), (5, "Arivu", "HR", 5500)]
emp_columns = ["emp_id", "name", "dept", "salary"]
df_emp = spark.createDataFrame(employees, emp_columns)
df_emp.show()
```

```python
departments = [("IT", "Chennai"),    ("HR", "Mumbai"), ("Finance", "Delhi")]
dept_columns = ["dept", "location"]
df_dept = spark.createDataFrame(departments, dept_columns)
df_dept.show()
```

```python
df_emp.select("name", "salary").show()
```

```python
df_emp.filter(df_emp.salary > 6000).show()
```

```python
from pyspark.sql.functions import avg

df_emp.groupBy("dept") \
    .agg(avg("salary").alias("avg_salary")) \
    .show()
```

```python
df_joined = df_emp.join(df_dept, on="dept", how="inner")
df_joined.show()
```

```python
df_emp.orderBy(df_emp.salary.desc()).show()
```
